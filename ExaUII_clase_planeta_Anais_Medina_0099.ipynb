```python
class Planeta:
    # El método __init__ define los atributos (datos)
    def __init__(self):
        self.nombre = ""
        self.masa = 0.0
        self.distancia_sol = 0.0

    # Función para leer datos desde el teclado
    def Leerdatos(self):
        print(f"--- Ingresando datos del planeta ---")
        self.nombre = input("Nombre del planeta: ")
        self.masa = input("Masa del planeta: ")
        self.distancia_sol = input("Distancia al sol (en millones de km): ")

    # Función para mostrar los datos almacenados
    def Mostrardatos(self):
        print(f"\n>>> Información de {self.nombre} <<<")
        print(f"Masa: {self.masa}")
        print(f"Distancia al sol: {self.distancia_sol} millones de km")
        print("-" * 30)

# --- Creación de los objetos (Instanciación) ---

# Creamos el primer objeto
planeta1 = Planeta()
planeta1.Leerdatos("", "", "")
planeta1.Mostrardatos()

# Creamos el segundo objeto
planeta2 = Planeta()
planeta2.Leerdatos("", "", "")
planeta2.Mostrardatos()
print("programa realizado por Anais medina NL = 36")
```
